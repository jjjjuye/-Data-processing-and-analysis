```python
import functools
li = [1,3,3,5,4,5]
print("sum: ", end = "") #end 안하면 줄 바꿈
print(functools.reduce(lambda a, b: a+b, li))

print("max:", end = "")
print(functools.reduce(lambda a, b : a if a>b else b, li)) #순서!!!  a if a>b else b
```

```python
import glob
all_files = glob.glob('**/myCabinetExcelData*.xls')

all_files
```

```python
import pandas as pd
all_files_data = []
for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)
    
all_files_data[0]
```

```python
all_files_data_concat = pd.concat(all_files_data, ignore_index = True)

all_files_data_concat
```
