```python
import os
import sys
import datetime
import time
import pandas as pd
import json
import urllib.request
def getRequestUrl(url):
    req = urllib.request.Request(url)
    try:
        response = urllib.request.urlopen(req)
        if response.getcode()==200:
            print("[%s] Url Request Success" %datetime.datetime.now())
            return response.read().decode('utf-8')
        
    except Exception:
            print("[%s] Error for URL : %s"%(datetime.datetime.now(), url))
            return None
        
        
def getTourismStatsItem():
    url = "http://openAPI.seoul.go.kr:8088/6e686f52686a796b36326e484f4346/json/ChunmanFreeSuggestions/1/1000"
    retData = getRequestUrl(url)
    
    if(retData == None):
        return None
    else:
        return json.loads(retData)


columns = [ 'TITLE', 'CONTENT', 'REG_DATE']
jsonData = getTourismStatsItem()


   
    
result = jsonData['ChunmanFreeSuggestions']['row']
rr = []

for i in range(1,1000):
    sn = result[i]['SN']
    title = result[i]['TITLE']
    content = result[i]['CONTENT']
    date = result[i]['REG_DATE']
    
    rr.append([sn,title, content, date])

   
    
    

columns = ['SN','TITLE','CONTENT','REG_DATE']
h=pd.DataFrame(rr, columns = columns)
h
dt=h.set_index('SN')
dt
```

```python
dt.to_csv("holly.csv", encoding = "utf-8", mode = "w")
```
